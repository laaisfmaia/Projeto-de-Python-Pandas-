## Criando novas variáveis

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
dados.head(10)
```

### 1) Variável que abrigará o valor bruto do aluguel

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

### 2) Variável que armazena o valor com base no metrô quadrado de um imóvel 

```python
dados['Valor m2'] = dados['Valor']/dados['Area']
dados
```

```python
dados['Valor m2'] = dados['Valor m2'].round(2)
```

```python
dados
```

### 3) Variável que abrigará o valor bruto do metro quadrado

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto']/dados['Area']).round(2)
```

```python
dados
```

### 4) Variável de tipo que agregue casas e apartamentos

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
dados
```

## Excluindo variáveis

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor m2', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
dados_aux
```
